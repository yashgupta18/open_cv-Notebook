```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

# Q1.HISTOGRAM EQUALIZATION

```python
fig1=cv2.imread("Figure1.tif")
F1=cv2.cvtColor(fig1,cv2.COLOR_BGR2GRAY)
plt.imshow(fig1)
plt.show
```

```python
H1=cv2.equalizeHist(F1) 
plot1=np.hstack((F1,H1)) 
plt.imshow(plot1)
```

```python
fig2=cv2.imread("Figure2.tif")
F2=cv2.cvtColor(fig2,cv2.COLOR_BGR2GRAY)
plt.imshow(fig2)
plt.show
```

```python
H2=cv2.equalizeHist(F2) 
plot2=np.hstack((F2,H2))
plt.imshow(plot2)
```

# Q2.RICE DETECTION

```python
fig3=cv2.imread("Figure3.tif")
F3=cv2.cvtColor(fig3,cv2.COLOR_BGR2HSV)
plt.imshow(fig3)
plt.show()
plt.imshow(F3)
plt.show()
```

```python
# define range of white color in HSV
white1=np.array([0,0,135])
white2=np.array([50,60,255])

# Threshold the HSV image to get only WHITE colors
mask = cv2.inRange(F3, white1, white2)

# Bitwise-AND mask and original image
res = cv2.bitwise_and(F3,F3, mask= mask)

plt.imshow(F3)
plt.show()
plt.imshow(mask)
plt.show()
plt.imshow(res)
plt.show()
```

```python
street=cv2.imread('color.tif')
plt.imshow(street)
```

```python
plt.hist(street.ravel(),256,[0,256]); plt.show()
```

```python
#bgr histogram
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([street],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
